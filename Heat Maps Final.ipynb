```python
import pandas as pd
```

```python
import json
import plotly.express as px
import requests


url = 'https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json'

response = requests.get(url)
counties = response.json()
```

```python
allMetroUSTop50 = pd.read_csv('Top 50 US Metro Final.csv')
allMetroUSTop50.head()
```

```python
rank_fig = px.choropleth(allMetroUSTop50, geojson=counties, locations='FIPS', color='Quantity',
 color_continuous_scale="mint",
 range_color=(200, 4500),
 scope='usa',
 labels={'Number of Households in County':'Number of CSV Files'}
 )
rank_fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
rank_fig.show()
```
