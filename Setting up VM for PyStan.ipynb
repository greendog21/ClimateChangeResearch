**Installation Guide** - All "" needs to be deleted inside the command line

Create a VM Instance


*   Follow this Video (https://www.youtube.com/watch?v=lIJlhKrP_SI) to create a VM Instance - Make Sure to Choose the computer that fullfills the requirements (For me the most basic computer works just takes a while)
*   Remember Everything is Linux Command Line Bassed - You will access the VM through SSH keys 
*   Use "wget https://repo.anaconda.com/archive/Anaconda3-2021.05-Linux-x86_64.sh" instead to install the newest version of python when the video uses wget

Transfering Files to the VM - Inside your Local Terminal


*   install gcloud - https://cloud.google.com/sdk/docs/install - Allows you to locally access/ talk to the VM through your own terminal
*   once installed do "gcloud init" and follow instructions | **NOTE** - Make sure your local username/user is the same as your GCP account user or else this will cause permission issues 
*   to transfer files like .py or .csv files to the VM use "gcloud compute scp *~*/path of local file/filename *~*/path of second local file/file name "name of your VM instance":*~*/narnia"
*   once you have transfered the python code over - install all the dependencies needed to run Stan/PyStan

Installing Dependencies On your Virtual Machine

*   After running "pip install pystan" try running the model
*   If package error first run "sudo apt-get update"
*   If still have an error then run "sudo apt-get install build-essential"












**Linux Command Guide**



*   ls - lists everything in the folder
*   rm [file name] - removes the file 
*   cd [folder name] - lets you select which folder to enter
*   cd .. - goes up a foldler/ back to the folder
*   cd / - goes to the root directory
*   python [python file] - runs the file once anaconda is installed
*   chmod [permissions] - changes the permisions of a file/directory
*   all Linux Commands - https://www.hostinger.com/tutorials/linux-commands
