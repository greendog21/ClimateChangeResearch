```python
import pandas as pd
```

```python
df = pd.read_csv('Top 50 US Metro Final.csv')
df.head()
```

```python
list1 = []
counter = 0 
for index, row in df.iterrows():
  counter += 1
  list1.append(counter)
```

```python
# plotting results
import matplotlib.pyplot as plt
```



1.   Outliers - Salt Lake City, Little Rock




```python
plt.figure(figsize=(65, 10))
plt.scatter( df['Geographic name'], df['Average Number of Occupants'], color='red', marker = 'o')
plt.xticks(range(0, 51, 1))
plt.ylabel('Average Number of Occupants', fontsize=14)
plt.xlabel('Metro Areas', fontsize=14)
plt.grid(True)
plt.show()
```



1.   No Outliers - Dallas has relatively greater floor area with lower real estate prices



```python
plt.figure(figsize=(65, 10))
plt.scatter( df['Geographic name'], df['Average Floor Area'], color='red', marker = 'o')
plt.xticks(range(0, 51, 1))
plt.ylabel('Average Floor Area (ft^2)', fontsize=20)
plt.xlabel('Metro Areas', fontsize=20)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(65, 10))
plt.scatter( df['Geographic name'], df['Housing Density'], color='red', marker = 'o')
plt.xticks(range(0, 51, 1))
plt.ylabel('Housing Density (F Area/ # of Occupants)', fontsize=14)
plt.xlabel('Metro Areas', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(65, 10))
plt.scatter( df['Geographic name'], df['Average Age of House'], color='red', marker = 'o')
plt.xticks(range(0, 51, 1))
plt.ylabel('Housing Density (F Area/ # of Occupants)', fontsize=14)
plt.xlabel('Metro Areas', fontsize=14)
plt.grid(True)
plt.show()
```
